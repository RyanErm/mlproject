## Incorporating COVID Data set 

We want to compare the number of COVID cases in different areas during 2020 - 2023 to the number of crimes against Asian individuals during that same time frame. 
We will use tables and graphs to do the comparison. 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

### Looking at the crime data set and trends

We are going to filter the crime data set to only include crimes against Asian individuals from 2020 - 2023.

```python
# loading in crime data frame 
crime_df = pd.read_csv('./cleaned_df.csv') 
crime_df = crime_df.drop(columns=['Unnamed: 0'])
crime_df.head()

# only want Year >= 2020, State, Month, VicRace == Asian
crime_df = crime_df[['State', 'Year','Month', 'VicRace']]
crime_df = crime_df[crime_df['Year'] >= 2020]
crime_df = crime_df[crime_df['VicRace'] == 'Asian']
crime_df = crime_df.reset_index(drop=True)
crime_df.head(20)

crime_df['State'].value_counts()
```

```python
import warnings
warnings.filterwarnings('ignore')

# adding another column to the data frame to indicate the month number, like how the covid data is done
crime_df['Month Number'] = 0
for i in range(len(crime_df)):
    # if jan, = 1
    if crime_df['Month'][i] == 'January':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 1
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 13
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 25
        else:
            crime_df['Month Number'][i] = 37

    # if feb = 2
    if crime_df['Month'][i] == 'February':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 2
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 14
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 26
        else:
            crime_df['Month Number'][i] = 38

    # if mar, = 3
    if crime_df['Month'][i] == 'March':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 3
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 15
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 27
        else:
            crime_df['Month Number'][i] = 39

    # if apr, = 4
    if crime_df['Month'][i] == 'April':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 4
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 16
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 28
        else:
            crime_df['Month Number'][i] = 40
    
    # if may, = 5
    if crime_df['Month'][i] == 'May':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 5
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 17
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 29
        else:
            crime_df['Month Number'][i] = 41

    # if jun, = 6
    if crime_df['Month'][i] == 'June':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 6
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 18
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 30
        else:
            crime_df['Month Number'][i] = 42

    # if jul, = 7
    if crime_df['Month'][i] == 'July':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 7
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 19
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 31
        else:
            crime_df['Month Number'][i] = 43

    # if aug, = 8
    if crime_df['Month'][i] == 'August':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 8
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 20
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 32
        else:
            crime_df['Month Number'][i] = 44

    # if sept, = 9
    if crime_df['Month'][i] == 'September':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 9
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 21
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 33
        else:
            crime_df['Month Number'][i] = 45

    # if oct, = 10
    if crime_df['Month'][i] == 'October':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 10
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 22
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 34
        else:
            crime_df['Month Number'][i] = 46

    # if nov, = 11
    if crime_df['Month'][i] == 'November':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 11
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 23
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 35
        else:
            crime_df['Month Number'][i] = 47

    # if dec, = 12
    if crime_df['Month'][i] == 'December':
        if crime_df['Year'][i] == 2020:
            crime_df['Month Number'][i] = 12
        elif crime_df['Year'][i] == 2021:
            crime_df['Month Number'][i] = 24
        elif crime_df['Year'][i] == 2022:
            crime_df['Month Number'][i] = 36
        else:
            crime_df['Month Number'][i] = 48   


crime_df.head(20)
```

```python
# 1 row of 4 graphs
fig, axes = plt.subplots(1, 4, figsize=(24, 6))  

# define years and month ranges
years = [2020, 2021, 2022, 2023]
month_ranges = [(1, 12), (13, 24), (25, 36), (37, 48)]  

# loop through years and corresponding axes
for ax, year, (start_month, end_month) in zip(axes, years, month_ranges):
    # filter data for each year
    filt = crime_df[crime_df['Month Number'].between(start_month, end_month)]
    
    # for each state
    for state in filt['State'].unique():
        state_df = filt[filt['State'] == state]
        # get the count of cases per month
        counts = state_df.groupby('Month Number').size()

        # plot
        ax.plot(counts.index - (start_month - 1), counts.values, label=state)

    # add title and labels 
    ax.set_title(f"Crimes with Asian Victims in {year}")
    ax.set_xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    ax.set_ylabel('Cases')
    ax.set_yticks(range(0, 7, 1))

# add legend
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, title="State", bbox_to_anchor=(1.0, 0.5), loc='center left')

plt.tight_layout()
plt.show()
```

### Looking at the COVID data set and trends

Filtering the COVID data set to only include cases from 2020 - 2023, in the states that the crime data set includes as well: California, Arizona, Alaska, Alabama, and Arkansas.

```python
# loading in data frame of covid monthly cases by state
covid_df = pd.read_csv('./cases_month_state.csv')
covid_df = covid_df.drop(columns=['Unnamed: 0'])
covid_df.head()

# CA = California, AZ = Arizona, AK = Alaska, AL = Alabama, AR = Arkansas
states = ['CA', 'AZ', 'AK', 'AL', 'AR']
covid_df = covid_df[covid_df['state'].isin(states)] 
covid_df = covid_df.reset_index(drop=True)
covid_df.head()
```

```python
# 1 row of 4 graphs
fig, axes = plt.subplots(1, 4, figsize=(24, 6))  

# define years and month ranges
years = [2020, 2021, 2022, 2023]
month_ranges = [(1, 12), (13, 24), (25, 36), (37, 48)]  

# loop through years and corresponding axes
for ax, year, (start_month, end_month) in zip(axes, years, month_ranges):
    # filter data for each year
    filt = covid_df[covid_df['month_number'].between(start_month, end_month)]
    
    # for each state
    for state in filt['state'].unique():
        state_df = filt[filt['state'] == state]
    
        # plot
        ax.plot(state_df['month_number'], state_df['new_cases'], label=state)

    # add title and labels 
    ax.set_title(f"COVID Cases per state in {year}")
    ax.set_xticks(range(start_month, end_month+1), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    ax.set_ylabel('Cases')

# add legend
handles, labels = axes[0].get_legend_handles_labels()
fig.legend( title="State", bbox_to_anchor=(1.0, 0.5), loc='center left')

plt.tight_layout()
plt.show()
```

### Plotting graphs next to each other and looking for comparisons 

```python
# 2 rows of 4 graphs
fig, axes = plt.subplots(2, 4, figsize=(24, 12))  
axes = axes.flatten()

# define years and month ranges
years = [2020, 2021, 2022, 2023]
month_ranges = [(1, 12), (13, 24), (25, 36), (37, 48)]  

# plot crime data in axes[0] to axes[3]
for i, (year, (start_month, end_month)) in enumerate(zip(years, month_ranges)):
    # set ax to the current axis
    ax = axes[i]

    # filter data for each year
    filt = crime_df[crime_df['Month Number'].between(start_month, end_month)]
    
    # for each state
    for state in filt['State'].unique():
        state_df = filt[filt['State'] == state]
        
        # get the count of cases per month
        counts = state_df.groupby('Month Number').size()

        # plot
        ax.plot(counts.index - (start_month - 1), counts.values, label=state)

    # add title and labels 
    ax.set_title(f"Crimes with Asian Victims in {year}")
    ax.set_xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    ax.set_ylabel('Cases')
    ax.set_yticks(range(0, 7, 1))

# plot covid data in axes[4] to axes[7]
for i, (year, (start_month, end_month)) in enumerate(zip(years, month_ranges)):
    # set ax to current axis 
    ax = axes[i + 4]

    # filter data for each year
    filt = covid_df[covid_df['month_number'].between(start_month, end_month)]
    
    # for each state
    for state in filt['state'].unique():
        state_df = filt[filt['state'] == state]
    
        # plot
        ax.plot(state_df['month_number'], state_df['new_cases'], label=state)

    # add title and labels 
    ax.set_title(f"COVID Cases per state in {year}")
    ax.set_xticks(range(start_month, end_month+1), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    ax.set_ylabel('Cases')

# adding legend 
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, title="States", bbox_to_anchor=(0.5, -0.05) , loc='lower center', ncol=6)

plt.tight_layout()
plt.show()
```
