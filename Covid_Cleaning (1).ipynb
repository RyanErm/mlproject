```python
import pandas as pd
```

```python
df = pd.read_csv("Weekly_Covid.csv")
df
```

```python
df = df.drop(['tot_deaths', "new_deaths", "new_historic_cases", "new_historic_deaths", "date_updated", "tot_cases", "start_date"], axis = 1)
```

we need to end up creating a new data frame where each row is the amount of covid cases for that month in that state in that year. Do this by

creating a new function that checks if the month is correct, and then the data is added is the amount of covid cases for that month. Create a new column that is date and new cases. Then in the function we can parse that string and look at the date and the cases.

```python
df['date'] = pd.to_datetime(df['end_date'], format='%m/%d/%Y')
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df
```

```python
#create a column representing the passage of time through months and years
# month number 1 is january 2020 and the final number is may 2023
df['month_number'] = ((df['year'] - 2020) * 12 + df['month'] - 1) + 1
```

```python
#create a dataframe for the new cases in each state in each month
#uses reset index to keep it as a dataframe, not a series
cases_state = df.groupby(['state', "month_number"])["new_cases"].sum().reset_index()
```
